# Customer Segmentation of Mall Customers Using K-Means Clustering

### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

### Data Collection and Analysis

```python
# Loading dataset into dataframe
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
# Get the number of rows in the dataset
df.shape[0]
```

```python
# Get information about the dataset
df.info()
```

```python
# Can see there are no null values, can also do the following:
df.isnull().sum()
```

### Choosing Annual Income and Spending Score Columns

```python
X = df.iloc[:,3:].values
```

### Choosing Number of Clusters:
#### WCSS - Within clusters sum of squares

```python
# Finding WCSS values for different numbers of clusters (1-10)
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, random_state=37)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
# Plot an elbow graph
sns.set()
plt.plot(range(1, 11), wcss)
plt.title('Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Optimal Number of Clusters = 5

### Training the K-Means Clustering Model

```python
kmeans = KMeans(n_clusters=5, init='k-means++', n_init=1, random_state=0)

# Return a label for each data point based on their cluster
Y = kmeans.fit_predict(X)
Y
```

### Visualizing the Clusters

```python
# Plotting the clusters and their centroids
plt.figure(figsize=(8,8))
plt.scatter(X[Y==0,0], X[Y==0,1], s=50, c='green', label='Cluster 1')
plt.scatter(X[Y==1,0], X[Y==1,1], s=50, c='red', label='Cluster 2')
plt.scatter(X[Y==2,0], X[Y==2,1], s=50, c='yellow', label='Cluster 3')
plt.scatter(X[Y==3,0], X[Y==3,1], s=50, c='violet', label='Cluster 4')
plt.scatter(X[Y==4,0], X[Y==4,1], s=50, c='blue', label='Cluster 5')

# Plot the centroids
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=150, c='black', label='Centroid')

plt.title('Customer groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
